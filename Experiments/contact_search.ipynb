```python
results = []
company="peepal"
query = f"site:linkedin.com/in/ 'Talent Acquisition' {company}"
res = search.results(query)

for result in res.get('organic', []):
      title = result.get('title', '').lower()
      snippet = result.get('snippet', '').lower()
      link = result.get('link')

      # Check if 'Talent Acquisition' present and company's name present in either title or snippet
      if link and 'talent acquisition' in title and company.lower() in (title + snippet):
          hr_name = result.get('title', '').split(' - ')[0].split(' at ')[0].strip()
          results.append({
              'Company': company,
              'HR Name': hr_name,
              'LinkedIn Profile': link
          })
```

```python
pprint.pprint(results)
```

```python
from openpyxl import load_workbook
output_df = pd.DataFrame(results)

# Save to Excel
output_file = "filtered_talent_acquisition_hr_profiles.xlsx"
output_df.to_excel(output_file, index=False)

# Load workbook to convert URLs to clickable links
wb = load_workbook(output_file)
ws = wb.active

# Write HYPERLINK formula for each link cell
for row in range(2, ws.max_row + 1):
    link_cell = ws[f'C{row}']
    url = link_cell.value
    if url:
        link_cell.value = f'=HYPERLINK("{url}", "{url}")'

# Save workbook
wb.save(output_file)
```
